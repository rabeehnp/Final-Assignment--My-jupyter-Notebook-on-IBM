```python
bitcoin_df['pct_change'] = bitcoin_df['Close'].pct_change()
bitcoin_df['pct_change'].fillna(0, inplace=True)
```

```python
sns.distplot(bitcoin_df['pct_change'])
```

## Volatality

```python
min_periods = 80
```

```python
volatality = bitcoin_df['pct_change'].rolling(min_periods).std() * np.sqrt(min_periods) 
volatality.fillna(0,inplace=True)
```

```python
volatality.plot()
```
